```python
import matplotlib.pyplot as plt
import numpy as np
import datetime
plt.rcParams.update({'font.size': 14})
```

# Import data

```python
# import data
mass = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", delimiter=",", usecols=[0, 1,2], skip_header=1, encoding="utf8")
# columns: total mass [g], tare [g], mass of the sample [g]

# time data
time = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", dtype='datetime64', delimiter=",", skip_header=1, usecols=[3], encoding="utf8")
time_in_seconds = time.astype('datetime64[s]').astype('int')
time_in_seconds = time_in_seconds - time_in_seconds[0] + 1

# count data
normalized_counts = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", delimiter=",", usecols=[8], skip_header=1, encoding="utf8")

print(mass[:,2])
```

# Background correction

```python
# decay-corrected counts
# note: each line is duplicated for easy subtraction from each vial
bg = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", delimiter=",", usecols=[5], skip_header=1, encoding="utf8")

normalized_counts = normalized_counts - bg

plt.title("Sample volume effect")
plt.plot(mass[:,2], normalized_counts, ".")
plt.xlabel("Mass of sample [g]")
plt.ylabel("Counts of 18F [CPM]")

plt.show()
```

# Relative efficiency

Compare the efficiency of counting against the 1 ml sample

```python
plt.title("Sample volume effect")
plt.plot(mass[:,2], normalized_counts/normalized_counts[0]*100, ".")
plt.xlabel("Mass of sample [g]")
plt.ylabel("Relative efficiency [%] \nCPM/(CPM @ 0.2 ml)")

plt.grid()
plt.show()
```

## This part is garbage but keeping for future ref

```python
concentration = normalized_counts/mass[:,2]

# find mass closest to 1 g
idx = (np.abs(mass[:,2] - 1)).argmin()

plt.title("Sample volume effect")
plt.plot(mass[:,2], concentration/concentration[idx], ".")
plt.xlabel("Mass of sample [g]")
plt.ylabel("Relative efficiency \n[CPM/ml]/[CPM/ml @ 1 ml] ")
#plt.hlines(1, -0.1, 4.5, linestyle="--", color="grey")
plt.grid()
plt.show()
```

```python
# combine the two

plt.title("Sample volume effect")
plt.plot(mass[:,2], normalized_counts/normalized_counts[0]*100, ".", label="13 mm")
plt.plot(mass_28[:,2], (normalized_counts_28)/max(normalized_counts_28)*100, ".", label="28 mm")
plt.xlabel("Mass of sample [g]")
plt.ylabel("Relative efficiency [%] \nCPM/(CPM @ 0.1 ml)")

plt.legend()
plt.grid()
plt.show()
```

# 28 mm data with extrapolation

```python
# import data
mass_28_extra = np.genfromtxt("Volumetric_Experiments/28_mm_volumetric_data_extrapolated.csv", delimiter=",", usecols=[0, 1,2], skip_header=1, encoding="utf8")
# columns: total mass [g], tare [g], mass of the sample [g]

# count data
normalized_counts_28_extra = np.genfromtxt("Volumetric_Experiments/28_mm_volumetric_data_extrapolated.csv", delimiter=",", usecols=[8], skip_header=1, encoding="utf8")

mass_28_extra = mass_28_extra[:,2]
mass_28_extra = mass_28_extra[np.isnan(mass_28[:,2])]

normalized_counts_28_extra = normalized_counts_28_extra[np.isnan(mass_28[:,2])]

# TODO: add background correction

plt.title("Sample volume effect")
plt.plot(mass[:,2], normalized_counts/normalized_counts[0]*100, ".", label="13 mm")
plt.plot(mass_28[:,2], (normalized_counts_28)/max(normalized_counts_28)*100, ".", label="28 mm")
plt.plot(mass_28_extra, (normalized_counts_28_extra)/max(normalized_counts_28)*100, "r.", label="28 mm, synthetic")
plt.xlabel("Mass of sample [g]")
plt.ylabel("Relative efficiency [%] \nCPM/(CPM @ 0.1 ml)")

plt.legend()
plt.grid()
plt.show()
```
