```python
import matplotlib.pyplot as plt
import numpy as np
import datetime
plt.rcParams.update({'font.size': 14})
```

## Note

The investigation of the sample volume effect was supposed to be done with two sets of vials: 13 mm diameter and 28 mm diameter, placed in the same kind of holders. However, the data for the 28 mm diameter vials seems to have an issue where the masses were not recorded for the second half of the experiment. 

# Import data

```python
# import data
mass = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", delimiter=",", usecols=[0, 1,2], skip_header=1, encoding="utf8")
# columns: total mass [g], tare [g], mass of the sample [g]

# time data
time = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", dtype='datetime64', delimiter=",", skip_header=1, usecols=[3], encoding="utf8")
time_in_seconds = time.astype('datetime64[s]').astype('int')
time_in_seconds = time_in_seconds - time_in_seconds[0] + 1

# count data
normalized_counts = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", delimiter=",", usecols=[8], skip_header=1, encoding="utf8")

print(mass[:,2])
```

# Background correction

```python
# decay-corrected counts
# note: each line is duplicated for easy subtraction from each vial
bg = np.genfromtxt("Volumetric_Experiments/13_mm_volumetric_data.csv", delimiter=",", usecols=[5], skip_header=1, encoding="utf8")

normalized_counts = normalized_counts - bg

plt.title("Sample volume effect")
plt.plot(mass[:,2], normalized_counts, ".")
plt.xlabel("Mass of sample [g]")
plt.ylabel("Counts of 18F [CPM]")

plt.show()
```

# Relative efficiency

Compare the efficiency of counting against the 1 ml sample

```python
plt.title("Sample volume effect")
plt.plot(mass[:,2], normalized_counts/normalized_counts[0]*100, ".")
plt.xlabel("Mass of sample [g]")
plt.ylabel("Relative efficiency [%] \nCPM/(CPM @ 0.2 ml)")

plt.grid()
plt.show()
```
